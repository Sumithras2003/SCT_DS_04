```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("road_accident_dataset.csv")
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Weather Conditions', 'Road Condition', 'Time of Day'])
```

```python
# Analyzing road condition
df['Road Condition'].value_counts().plot(kind='bar', title='Accidents by Road Condition')
plt.show()
```

```python
# Analyzing weather condition
df['Weather Conditions'].value_counts().plot(kind='bar', title='Accidents by Weather')
plt.show()
```

```python
# Analyzing time
df['Time of Day'].value_counts().plot(kind='bar', title='Accidents by Time of Day')
plt.show()
```

```python
#Accident Hotspots by Region
sns.countplot(data=df, y='Region', order=df['Region'].value_counts().index, color='skyblue')
plt.title('Accidents by Region')
plt.xlabel('Number of Accidents')
plt.ylabel('Region')
plt.show()
```

```python
pivot = df.pivot_table(index='Weather Conditions', columns='Time of Day', aggfunc='size', fill_value=0)
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Weather vs Time of Day - Accident Count')
plt.show()

```

```python
# Contributing factors
cause_counts = df['Accident Cause'].value_counts()
cause_counts.plot(kind='barh', title='Top Accident Causes', color='salmon')
plt.xlabel('Number of Accidents')
plt.ylabel('Accident Cause')
plt.show()
```
